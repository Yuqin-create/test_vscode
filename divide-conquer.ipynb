```python
#binary search
def binary_search(sorted_A,m):
    N=len(sorted_A)
    if N==1:
        if sorted_A[0]==m:
            print(f"{m} is found")
        else:
            print(f"{m} is not found")
    elif sorted_A[int(N/2)]== m: 
        print(f"{m} is found in {sorted_A}")
    elif sorted_A[int(N/2)]> m:
        sorted_B=sorted_A[:int(N/2)]
        binary_search(sorted_B,m)
    elif sorted_A[int(N/2)]< m:
        sorted_B=sorted_A[int(N/2):]
        binary_search(sorted_B,m)
```

```python
A=[3,4,7,9,12,17,29,89]
M=50
N=binary_search(A,M)
```

```python
#fibonacci_num
import numpy as np
def fibonacci_num(n):
    if n==0:
        Fn=0
        return Fn
    if n==1:
        Fn=1
        return Fn
    else:
        A=np.array([[1,1],[1,0]])
        P=powern(A,n-1)
        Fn=P[0,0]
        return Fn
###
def powern(A,n):
    if n==1:
        return A
        
    elif n%2==0:
        P=powern(A,n/2)
        return P@P
    else:
        P=powern(A,(n-1)/2)
        return P@P@A
```

```python
Fn=fibonacci_num(3)
print(Fn)
```

```python
#Maximum-subarray Use the array A = [-2, 11, -4, 13, -5, -2] to test your code.
def max_subarray_cross(A):
    n=len(A)
    mid=n//2
    right_sum=-np.inf
    sum=0
    for i in range(mid,n):
        sum+=A[i]
        if sum>right_sum:
            right_sum=sum
            max_right=i
    left_sum=-np.inf
    sum=0
    for i in range(mid-1,0,-1):
        sum+=A[i]
        if sum>left_sum:
            left_sum=sum
            max_left=i
    return(right_sum+left_sum)
        
def max_subarray(A):
    n=len(A)
    if n==1:
        return A[0]
    else:
        mid=n//2
        right=max_subarray(A[:mid])
        left=max_subarray(A[mid:])
        cross=max_subarray_cross(A)
        if right>left and right>cross:
            return right
        elif left>right and left>cross:
            return left
        else:
            return cross
    
```

```python
#Strassenâ€™s algorithm for matrix multiplication
import numpy as np
def strassen(A,B):
    n=A.shape[0]
    if n==1:
        return A[0,0]*B[0,0]
    elif n%2==0:
        A11=A[:n//2,:n//2]
        A12=A[:n//2,n//2:]
        A21=A[n//2:,:n//2]
        A22=A[n//2:,n//2:]
        B11=B[:n//2,:n//2]
        B12=B[:n//2,n//2:]
        B21=B[n//2:,:n//2]
        B22=B[n//2:,n//2:]
        S1=B12-B22
        S2=A11+A12
        S3=A21+A22
        S4=B21-B11
        S5=A11+A22
        S6=B11+B22
        S7=A12-A22
        S8=B21+B22
        S9=A11-A21
        S10=B11+B12
        P1=strassen(A11,S1)
        P2=strassen(S2,B22)
        P3=strassen(S3,B11)
        P4=strassen(A22,S4)
        P5=strassen(S5,S6)
        P6=strassen(S7,S8)
        P7=strassen(S9,S10)
        C11=P5+P4-P2+P6
        C12=P1+P2
        C21=P3+P4
        C22=P5+P1-P3-P7
        C=np.block([[C11,C12],[C21,C22]])
        return C   
    elif n%2==1:
        A=pad_matrix(A)
        B=pad_matrix(B)
        T=strassen(A,B)
        m=T.shape[0]
        C=T[:m-1,:m-1]
        return C
        
 ####
def pad_matrix(A):
    A=np.vstack((A,np.zeros((1,A.shape[1]))))
    A=np.hstack((A,np.zeros((A.shape[0],1))))
    return A
        
```

```python
#Randomize quick sort
import numpy as np
def Randomize_quick_sort(A):
    n=len(A)
    if n<=3:
        return A
    else:
        r=np.random.randint(0,n)
        t=A[0]
        A[0]=A[r]
        A[r]=t
        i=0
        for j in range(1,n):
            if A[j]<=A[0]:
                i+=1
                t=A[i]
                A[i]=A[j]
                A[j]=t
        t=A[0]
        A[0]=A[i]
        A[i]=t
        R=Randomize_quick_sort(A[:i])
        L=Randomize_quick_sort(A[i+1:])
        S=np.hstack((R,A[i],L))
        return S
```
