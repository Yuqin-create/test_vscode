```python
#max heapify
class myheap: 
    def __init__(self,value):
        self.value=value
        self.length=len(value)
        self.heapsize=len(value)
    def  __str__(self):
        return str(self.value)
def Maxheapify(A,i):
    l=2*i
    r=2*i+1
    if (l<=A.heapsize) and (A.value[l-1]>A.value[i-1]):
        largest=l
    else:
        largest=i
    if (r<=A.heapsize) and (A.value[r-1]>A.value[largest-1]):
        largest=r
    if largest!=i:
        A.value[i-1], A.value[largest-1] = A.value[largest-1], A.value[i-1]
        Maxheapify(A,largest)
    return A
```

```python
import math
B=[16,4,10,14,9,3,2,8,1]
H=myheap(B)

MaxH=Maxheapify(H,2)

print(MaxH)
        
```

```python
#build max heap
def build_maxheap(A):
    for i in range(A.length//2,1,-1):
        Maxheapify(A,i)
    return A
```

```python
B=[16,4,10,14,9,3,2,8,1]
H=myheap(B)
MaxH=build_maxheap(H)
print(MaxH)
        
```

```python
#heapsort
def heapsort(A):
    for i in range(A.length,1,-1):
        A.value[i-1], A.value[0] = A.value[0], A.value[i-1]
        A.heapsize-=1
        Maxheapify(A,1)
    return A
```

```python
B=[16,4,10,14,9,3,2,8,1]
H=myheap(B)
MaxHS=heapsort(H)
print(MaxHS)
```
