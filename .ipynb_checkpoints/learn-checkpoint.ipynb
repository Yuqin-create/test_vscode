```python
def pad_matrix(A):
    A=np.vstack((A,np.zeros((1,A.shape[1]))))
    A=np.hstack((A,np.zeros((A.shape[0],1))))
    return A
T=np.array([[1,0,0],[0,1,0],[0,0,1]])
m=T.shape[0]
C=T[:m-1,:m-1]
print(C)
T=pad_matrix(C)
m=T.shape[0]
C=T[:m-1,:m-1]
print(C)
```

```python
import numpy as np
#count sort
def countsort(A,EXP):
    C=np.zeros(10,dtype=int)
    n=len(A)
    B=np.zeros(n,dtype=int)
    digits=(A//EXP)%10
    for i in digits:
        C[i]+=1
    for j in range (1,10):
        C[j]+=C[j-1]
    for m in range (n-1,-1,-1):
        B[C[digits[m]]-1]=A[m]
        C[digits[m]]-=1
    return B
#radix sort
def radixsort(A):
    A=np.array(A,dtype=float)
    minV=np.min(A)
    if minV<0:
        A+=abs(minV)
    max_decimal=0
    for num in A:
        if num%1 !=0:
            decimal_str=str(num).split('.')[1]
            max_decimal=max(max_decimal,len(decimal_str))
    if max_decimal>0:
        A=np.round(A*10**max_decimal).astype(int)
    else:
        A=A.astype(int)
    maxV=np.max(A)
    exp=1
    while maxV//exp>0:
        A=countsort(A,exp)
        exp*=10   
    if max_decimal>0:
        A=A/10**max_decimal
    if minV<0:
        A-=abs(minV)
    return A

A=np.array[4,1,3,4,3,-0.9,-1]
countsort(A,1)
radixsort(A)
```
